# K Nearest Neighbor (from scratch)

```python
from sklearn.datasets import load_iris
import numpy as np
import random
```

```python
my_data = load_iris()           # Loading data using sklearn library
print(my_data)

data = my_data.data             # Fetching data
#print(my_data.data)

x1 = data[:,0]
x2 = data[:,1]
x3 = data[:,2]
x4 = data[:,3]
```

```python
# Defining euclidian distance so as to select the minimum distances for comparison.
def euc_distance(clusters,p,k):
              distance = []
              for cluster in clusters:  # this will take one class from clusters either 0 or 1 or 2.
                            for tup in clusters[cluster]: 
                                          # this will take tuples one by one from the class selected above
                                          eucl_dist = ((tup[0]-p[0])**2+(tup[1]-p[1])**2+(tup[2]-p[2])**2+(tup[3]-p[3])**2)**(1/2)
                                          # calculating the euclidian distance between each tuple
                                          # of selected class and the point taken at random
                                          distance.append((eucl_dist,cluster))
                                          # distances are appended with the respective class
                                         
              distance = sorted(distance)[:k]   # sorting the list of calculated distances for 
                                                # taking k number of minimum distances.
              class_0 = 0
              class_1 = 0
              class_2 = 0
              for i in distance:
                            if i[1]==0:      # here, checking whether the class of the selected distance
                                             # is 0 or 1 or 2
                                          class_0 += 1
                            elif i[1]==1:
                                          class_1 += 1
                            else:
                                          class_2 += 1

              if class_0>class_1:
                            if class_0>class_2:
                                          return 0
              elif class_1>class_2:
                            return 1
              else:
                            return 2


zipped_data = list(zip(x1,x2,x3,x4)) # Here, we are zipping our data to have it in the 
                                     # form of tuples in one list
#print(zipped_data)
```

```python
##TRAINING DATA##.....70%.....
# Splitting the data without sklearn, here, test_size=0.3
# 30% data will be tested and the remaining come under training data
# data is splitted into 3 clusters i.e., class_0, class_1 and class_2
clusters = {0: zipped_data[:35], 1: zipped_data[50:85], 2: zipped_data[100:135]}

##TESTING DATA##
p = (zipped_data[35:50]+zipped_data[85:100]+zipped_data[135:150])

k = 5
# k is the number of nearest neighbours.
# k=5 refers to the number of minimum distances that are then used for selecting k minimum distances.
'''for i in range(len(p)):
              p = random.choice(p)
              print('for point',p,'the class to which it belongs:',euc_distance(clusters,p,k))'''
```

```python
##ACTUAL RESULT##
# collecting the actual result
points = []
y_actual =[]
for i in range(50):
              point = random.choice(p)
              points.append(point)
              if p.index(point)<25:
                            y_actual.append(0)
              elif (p.index(point)>=25 and p.index(point)<50):
                            y_actual.append(1)
              else:
                            y_actual.append(2)

#print(points)
```

```python
##PREDICTED RESULT##
y_pred = []
accuracy = 0
for i in points:
              y_pred.append(euc_distance(clusters,i,k))
```

```python
##ACCURACY##
for i in range(50):
              if y_pred[i]==y_actual[i]:
                            accuracy = accuracy+1
                            

print('accuracy of the result:',((accuracy/50)*100),'%')
```
