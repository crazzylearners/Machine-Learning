## PairGrid(data, hue=None, hue_order=None,pallete=None,vars=None,
## x_vars=None,y_vars=None,aspect=1,height=2.5,size=None)

```python
iris=sns.load_dataset('iris')
iris
```

```python
x=sns.PairGrid(iris)
y=x.map(plt.scatter)
y
```

```python
x=sns.PairGrid(iris)
y=x.map_diag(plt.hist)
y=x.map_offdiag(plt.scatter)
y
```

```python
x=sns.PairGrid(iris,hue='species')
y=x.map(plt.scatter)
y.add_legend()
y
```

```python
x=sns.PairGrid(iris,hue='species',vars=['sepal_length','sepal_width'])
y=x.map(plt.scatter)
y.add_legend()
y
```

```python
x=sns.PairGrid(iris,hue='species')
y=x.map(sns.regplot)
y.add_legend()
y
```

```python
x=sns.PairGrid(iris,hue='species',hue_kws={'marker':['d','s','+']})
y=x.map(plt.scatter)
y.add_legend()
y
```

## FacetGrid(data,row=None,col=None,hue=None,hue_order=None,row_order=None,
##  col_order=None,xlim=None,ylim=None,hue_kws=None,aspect=1,height=3,size=None)

```python
x=sns.FacetGrid(tip,row='smoker', col='time')
y=x.map(plt.hist, 'total_bill')
y
```

```python
x=sns.FacetGrid(tip,row='smoker', col='time')
y=x.map(plt.scatter, 'total_bill','tip')
y
```

```python
x=sns.FacetGrid(tip,hue='smoker', col='time')
y=x.map(plt.scatter, 'total_bill','tip')
y.add_legend()
```

```python
x=sns.FacetGrid(tip,row='smoker', col='time')
y=x.map(sns.regplot, 'total_bill','tip')
y
```

## kdeplot()

```python
mean=[0,0]
cov=[[0.2,0],[0,3]]
x_axis,yaxis=np.random.multivariate_normal(mean,cov,size=40).T
sns.kdeplot(x_axis)
```

```python
mean=[0,0]
cov=[[0.2,0],[0,3]]
x_axis,y_axis=np.random.multivariate_normal(mean,cov,size=40).T
sns.kdeplot(y_axis)
```

```python
mean=[0,0]
cov=[[0.2,0],[0,3]]
x_axis,yaxis=np.random.multivariate_normal(mean,cov,size=40).T
sns.kdeplot(x_axis,shade=True, color='green')
```

```python
sns.kdeplot(x_axis,y_axis)
```

```python
sns.kdeplot(x_axis,y_axis,shade=True)
```

```python
sns.kdeplot(x_axis,y_axis,cmap='RdBu')
```

```python
sns.kdeplot(x_axis,y_axis,n_levels=35,cmap='coolwarm')
```

```python
sns.kdeplot(y_axis,bw=1.5)
```

```python
sns.kdeplot(y_axis)
```
